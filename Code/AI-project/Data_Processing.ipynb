```python
import pandas as pd

df = pd.read_csv('Edited Dataset.csv', engine='python', on_bad_lines='warn')
differentJobs = []
jobCounter = {}
uniqueIndex = {}

for i in df['job_description_text']:
    if i not in differentJobs:
        differentJobs.append(i)
        jobCounter[differentJobs.index(i)] = 1
        uniqueIndex[differentJobs.index(i)] = i
    else:
        jobCounter[differentJobs.index(i)] += 1

refined_data = {
    k: v for k, v in jobCounter.items() if v >= 10
}
```

```python
refined_jobs_set = set(uniqueIndex[k] for k in refined_data.keys())

refined_df = df[df['job_description_text'].isin(refined_jobs_set)]
```

```python
refined_df
```

```python
refined_df['job_id'] = refined_df['job_description_text'].factorize()[0]
```

```python
refined_df
```

```python
# Keep only jobs with at least one accepted and one rejected CV
valid_jobs = refined_df.groupby('job_id')['decision'].nunique()
valid_jobs = valid_jobs[valid_jobs > 1].index
refined_df = refined_df[refined_df['job_id'].isin(valid_jobs)]
```

```python
refined_df
```

```python
refined_df.to_csv('Edited Dataset.csv', index=False)
```
